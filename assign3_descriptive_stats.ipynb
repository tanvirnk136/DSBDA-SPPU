```python
def calculate_mean(data):
    return sum(data)/len(data)

def calculate_std(data,mean):
    variance=sum((x-mean)**2 for x in data)/ (len(data)-1)
    return variance**0.5

def calculate_median(data):
    sorted_data=sorted(data)
    n=len(sorted_data)
    if n%2==0:
        return (sorted_data[n//2-1]+ sorted_data[n//2])/2
    else:
        return sorted_data[n//2]

def calculate_min(data):
    return min(data)

def calculate_max(data):
    return max(data)

groups=df.groupby('income')
for income, group in groups:
    age_data=group['age']
    mean_age=calculate_mean(age_data)
    std_age=calculate_std(age_data,mean_age)
    median_age= calculate_median(age_data)
    min_age=calculate_min(age_data)
    max_age=calculate_max(age_data)

print(f" Statistics for income group'{income}':")
print(f" Mean age: {mean_age}")
print(f" Standard deviation of age: {std_age}")
print(f" Median age: {median_age}")
print(f" Min age:{min_age}")
print(f" Max age:{max_age}")
print()
    
```

```python
def displayStats(data,name):
    print("-----------Statistics for", name, "species set-----------")
    print("\nCount:\n", data.select_dtypes(include='number').count())
    print("\nMean:\n", data.select_dtypes(include='number').mean())
    print("\nMedian:\n", data.select_dtypes(include='number').median())
    print("\nMin:\n", data.select_dtypes(include='number').min())
    print("\nMax:\n", data.select_dtypes(include='number').max())
    print("\nStandard Deviation:\n", data.select_dtypes(include='number').std())
    print("\n25%:\n", data.select_dtypes(include='number').quantile(0.25))
    print("\n75%:\n", data.select_dtypes(include='number').quantile(0.75))

displayStats(s,'Setosa')
displayStats(vc,'Verginica')
displayStats(s,'Versicolor')
```

```python
import math
def average(df, col):
    avg=sum(df[col])/len(df[col])
    print(col, ":", avg)
def standardDev(df, col):
    m=df[col].mean()
    summation=0
    for i in df[col]:
        summation+=(i-m)**2
    std=math.sqrt(summation/(len(df[col])-1))
    print(col, ":", std)
def Median(data, col):
    n_list=list(data[col])
    n = len(data[col])
    n_list.sort()
    if n % 2 == 0:
        m1 = n_list[n//2]
        m2 = n_list[n//2 - 1]
        median = (m1 + m2)/2
    else:
        median = n_list[n//2]
    print(col, ":", median)
def percentile(data, col, p):
    n_list=list(data[col])
    n = len(data[col])
    n_list.sort()
    index = int(p * n)
    print(col, ":", n_list[index])
def displayStats(data, name):
    numeric_columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
    print("--------------------------Statistics for", name, "species set-------------------------")
    print("\nMean: ")
    for col in numeric_columns:
        average(data, col)
    print("\nStandard Deviation: ")
    for col in numeric_columns:
        standardDev(data, col)
    print("\n25th Percentile: ")
    for col in numeric_columns:
        percentile(data, col, 0.25)
    print("\nMedian: ")
    for col in numeric_columns:
        Median(data, col)
    print("\n75th Percentile: ")
    for col in numeric_columns:
        percentile(data, col, 0.75)


displayStats(s, "Setosa")
print()
displayStats(vc, "Versicolor")
print()
displayStats(vg, "Virginica")
```
