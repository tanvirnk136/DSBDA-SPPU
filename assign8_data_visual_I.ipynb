```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("titanic")
df
```

```python
plt.figure(figsize=(3, 4))
sns.histplot(data=df,x="fare") #Both parameters can be continuous
```

```python
sns.histplot(data=df,x="fare",binwidth=30)
```

```python
sns.histplot(data=df,x="fare", binwidth=30,hue="sex") # Hue - gives data in terms of another attribute
#Fare of Class 1 people is less and as the class increases the Fare also increases
```

```python
sns.displot(df["fare"], aspect=2.5)
```

```python
sns.displot(df["fare"], aspect=2.5, bins=20)
```

```python
sns.displot(data=df,x="age", hue="sex") # Hue is categorical data
# Overall Male passengers have more count than Female
# Female passengers are more in count between age 5 to 15
# No Female passengers above the age 65
```

```python
sns.displot(data=df,x="age", hue="sex",col="pclass")
# Huge count of Males in Class 3
# No Females in Class 3 above 47 age
# Most number of Females in class 1 between the age of 15 to 45
# Class 2 has no passengers between age 7 to 12
```

```python
sns.displot(data=df,x="age", hue="sex", col="pclass",kde=True)
```

```python
sns.displot(data=df, x="age", hue="alive")
# Children between age 0 to 10 survived
# People with age greater than 77 survived
# People with age 20-30 couldn't survive
```

```python
sns.displot(data=df,x="sex",hue="alive")
# More number of Females survived as compared to Males
# Very less Females couldn't survive
# Very less Males could survive
```

```python
sns.barplot(x='sex', y='age', data=df) # 1 categorical and other continuous
# Average age of Male passengers is 30 approx.
# Average age of Female passengers is 25-30
```

```python
sns.barplot(x='sex', y='age', data=df, hue="alive")
```

```python
sns.countplot(x='pclass', data=df) # One attribute is categorical
# Class 3 passengers are most in number
# Class 2 passengers are least in number
```

```python
sns.barplot(x="pclass", y="age", data=df, hue="embarked", ci=None)
# Class 1 - Age 30 to 35, embarked from all cities
# Class 2 - Age 40+, embarked from Queenstown
# Class 3 - Age 20 to 30, embarked from Southampton and Queenstown
```

```python
sns.countplot(x='pclass', data=df, hue="sex")
# Ratio of Males to Females is more than Double
```

```python
sns.countplot(x='pclass', data=df, hue="sex", dodge=False)
```

```python
sns.catplot(x="embarked", data=df, hue="survived", kind="count") # X is categorical
# Most number of Passengers who died, boarded from Southampton
# More passengers survived from Cherbourg city
```

```python
sns.catplot(x="embarked", data=df, hue="survived", kind="count", col="pclass")
# Class 1 - More survivers from Southampton and Cherbourg compared with Death Toll
# Very less passengers from Queenstown
# Class 2 - More survivers from Cherbourg comparing Death Toll
# Class 3 - Most people in Class 3 died
# Death Toll is more than survived
# Most people who died belonged to Southampton
```

```python
sns.displot(data=df, x="fare", hue="survived", binwidth=40)
# People with less fare couldn't survive
```

```python
sns.catplot(x="alone", data=df, hue="survived", kind="count")
# People who were Alone couldn't survive
# Suvival rate was more if person was not Alone
```

```python
sns.countplot(x='deck', data=df, hue="survived")
# Survival rate of Deck A and G is NOT good
# Rest all decks had more number of Survivals
```
