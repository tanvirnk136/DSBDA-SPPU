```python
# Outliers are only present in sepal_width column
sw = df[['sepal_width','species']]
```

```python
Q1= sw['sepal_width'].quantile(0.25)
Q3 = sw['sepal_width'].quantile(0.75)
IQR = Q3-Q1
print("Q1: ",Q1," | Q3: ",Q3," | IQR: ",IQR)
```

```python
LB=Q1-1.5*IQR
UB=Q3+1.5*IQR
print("LB: ",LB," | UB: ",UB)
```

```python
for i in sw['sepal_width']:
    if(i<LB or i>UB):
        print(i)
```
