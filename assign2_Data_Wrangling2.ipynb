```python
#outlier detection using visualization(boxplot)
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
nc=['Subject 1','Subject 2','Subject 3','Subject 4','Attendance']
for col in nc:
    sns.boxplot(x=df[col])
    plt.title(f'boxplot of{col}')
    plt.show()
```

```python
from scipy import stats

# Initialize an empty column for outliers
df['IsOutlier'] = False

# Set a threshold (e.g., 2)
threshold = 2

for col in nc:
    # Calculate Z-scores for the current column
    z_scores = stats.zscore(df[col])
    
    # Identify outliers and update 'IsOutlier' column
    df['IsOutlier'] = df['IsOutlier'] | (np.abs(z_scores) > threshold)

# Display the DataFrame
print(df)
```

```python
#HAndling Outliers using IQR
for col in nc:
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR=Q3-Q1
    lb=Q1-1.5*IQR
    ub=Q3+1.5*IQR
    df[col]=np.where((df[col]<lb) | (df[col]>ub),df[col].median(),df[col])
df
```
