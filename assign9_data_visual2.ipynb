```python
df = df.drop(['deck'],axis=1)
```

```python
Q1=df['age'].quantile(0.25)
Q3=df['age'].quantile(0.75)
IQR=Q3-Q1
print("Q1: ",Q1," | Q3: ",Q3," | IQR: ",IQR)
```

```python
LB=Q1-1.5*IQR
UB=Q3+1.5*IQR
print("LB: ",LB," | UB: ",UB)
```

```python
for i in df['age']:
    if(i>UB):
        print(i)
```
