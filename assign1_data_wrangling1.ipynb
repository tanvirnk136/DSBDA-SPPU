```python
#make calories value absolute (i.e. remove the negative sign of all negative values)
df['Calories']=df['Calories'].abs()
df.head(10)
```

```python
#show NULL values of columns 
df.isnull().sum()
```

```python
# replace null with average in Calories
df['Calories'].fillna(df['Calories'].mean(),inplace=True)
df.head(10)
```

```python
df.dropna(subset=['Date'],inplace=True)
df.head(10)
```

```python
# changes value of 7th record from 450 to 45
df.loc[7, 'Duration']=45
df.head(10)
```

```python
# convert calories to int data type
df['Calories']=df['Calories'].astype(int)
df.head(10)
```

```python
# converts into datetime data-type
df['Date']=pd.to_datetime(df['Date'])
df.head(10)
```

```python
from sklearn import preprocessing
```

```python
data_nba=pd.read_csv('C:/Users/TANVI RAINAK/Downloads/nba.csv')
data_nba
```

```python
# gives unique values of Position column
data_nba['Position'].unique()
```

```python
# label encoder
data = preprocessing.LabelEncoder()
```

```python
#transforms position column
data_nba['Position'] = data.fit_transform(data_nba['Position'])
data_nba
```

```python
# makes groups according to age
x=pd.cut(data_nba.Age, bins=[19,25,30,35,45],labels=['A','B','C','D'])
```

```python
# inserts another column at 5th position with name Age_Group
data_nba.insert(5,'Age_group',x)
data_nba
```
